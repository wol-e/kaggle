```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/train_folds.csv")
```

```python
asset_details = pd.read_csv("../data/raw/asset_details.csv")
asset_details
```

```python
weights = asset_details.set_index("Asset_ID", drop=True)[["Weight"]]
weights
```

```python
df = df.join(weights, on="Asset_ID", how="left")
df
```

```python
import sys
sys.path.append("..")

from src.helpers.memory import reduce_memory_usage
```

```python
df = reduce_memory_usage(df, print_stats=True)
```

```python
df = df.sort_values(by=["timestamp", "Asset_ID"]).reset_index(drop=True)
df
```

```python
first_opens = {}
for asset_id in asset_details.Asset_ID:
    first_open = df[df["Asset_ID"] == asset_id].Open.values[0]
    first_opens[asset_id] = first_open
    
first_opens
```

```python
asset_id = 0
field = "Close"
df.loc[df["Asset_ID"] == asset_id,field]
df[df["Asset_ID"] == asset_id][field] / first_opens[asset_id]
```

```python
from tqdm.notebook import tqdm


for asset_id in tqdm(asset_details.Asset_ID):
    for field in ["Open", "High", "Low", "Close"]:
        df.loc[df["Asset_ID"] == asset_id,field] = df[df["Asset_ID"] == asset_id][field] / first_opens[asset_id]
        
df
```

```python
df = df[["timestamp", "Count", "Open", "High", "Low", "Close", "Weight"]]
df
```

```python
for field in ["Open", "High", "Low", "Close"]:
    df[field] = df[field] * df["Weight"]
    
df
```

```python
df = df.drop("Weight", axis=1)
```

```python
df = df.groupby("timestamp").agg({
    "Count": ["sum"],
    "Open": ["sum"],
    "High": ["sum"],
    "Low": ["sum"],
    "Close": ["sum"]
})
```

```python
df.columns = ["Market_Count", "Market_Open", "Market_High", "Market_Low", "Market_Close"]

df = df.reset_index()

df
```

```python
df.to_csv("../data/processed/market.csv", index=False)
```
